#Read Data

```python

import json
import pandas as pd

# Initialize an empty list to hold all the data
all_data = []

# Read the data from data.txt
with open('detObj2.txt', 'r') as file:
    for line in file:
        # Parse each line as JSON
        data = json.loads(line.strip())
        
        # Check if 'time' and 'numObj' are in the data
        if 'time' not in data or 'numObj' not in data:
            continue  # Skip this entry if either key is missing
        
        time = data['time']
        numObj = data['numObj']
        
        # Iterate over the number of objects and construct a row for each object
        for i in range(numObj):
            row = {
                'time': time,
                'numObj': numObj,
                'x': data['x'][i] if 'x' in data else None,
                'y': data['y'][i] if 'y' in data else None,
                'z': data['z'][i] if 'z' in data else None,
                'doppler': data['doppler'][i] if 'doppler' in data else None,
                'snr': data['snr'][i] if 'snr' in data else None,
                'noise': data['noise'][i] if 'noise' in data else None
            }
            all_data.append(row)

# Convert the list of rows to a DataFrame
final_df = pd.DataFrame(all_data)

# print(final_df)
df =final_df
```

```python
import os
os.environ["KMP_INIT_AT_FORK"] = "FALSE"  # Disable OpenMP threading

import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D


# Ensure the DataFrame is sorted by time
df['time'] = pd.to_datetime(df['time'], unit='s')
df = df.sort_values('time')

# Group the data by unique timestamps
grouped = df.groupby('time')
unique_times = list(grouped.groups.keys())

# Create the figure and 3D axis
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Motion Over Time')

# Add a text annotation for the timestamp
time_text = ax.text2D(0.02, 0.95, '', transform=ax.transAxes)

# Define a color map for clusters
n_clusters = 3
colors = plt.cm.get_cmap('viridis', n_clusters)

# Animation update function
def update(frame):
    ax.clear()
    
    # Set the axis limits and labels again (since we cleared the axis)
    ax.set_xlim(df['x'].min() - 1, df['x'].max() + 1)
    ax.set_ylim(df['y'].min() - 1, df['y'].max() + 1)
    ax.set_zlim(df['z'].min() - 1, df['z'].max() + 1)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('3D Motion Over Time')
    
    # Get data for the current timestamp
    current_time = unique_times[frame]
    data = grouped.get_group(current_time)
    
    # Perform KMeans clustering if there are enough data points
    if len(data) >= n_clusters:
        kmeans = KMeans(n_clusters=n_clusters, random_state=0)
        data['cluster'] = kmeans.fit_predict(data[['x', 'y', 'z']])
    else:
        data['cluster'] = [0] * len(data)  # Assign all points to a single cluster
    
    # Plot all points for the current timestamp, colored by cluster
    scatter = ax.scatter(data['x'], data['y'], data['z'], c=data['cluster'], cmap=colors, s=50, alpha=0.5)
    
    # Update the timestamp text
    time_text.set_text(f'Time: {current_time}')
    
    return scatter, time_text

# Create the animation
anim = FuncAnimation(fig, update, frames=len(unique_times), interval=500, blit=False)

# Show the animation
plt.show()

# Save the animation as a GIF
try:
    anim.save('3d_motion_animation.gif', writer='pillow', fps=2)
except Exception as e:
    print(f"Error saving animation: {e}")
```
